```python
import pandas as pd
```

```python
data = pd.read_csv("./data/archive/winequality-red.csv")
```

```python
# Заменяме интервал с долна черта в иената на колоните, защото иначе дава грешка при запис на XML
for col in data.columns:
    if " " in col:
        new_col = col.replace(" ", "_")
        data.rename(columns={col: new_col}, inplace=True)
```

```python
# indent=4 подрежда по 4 интервала
data.to_json("C:/Users/user/Downloads/data.json", orient="records", indent=4)
```

```python
data.rename(columns={"fixed acidity": "fixed_acidity_"}, inplace=True)
data.to_xml("C:/Users/user/Downloads/data.xml", index=False, encoding="utf-8")
```
