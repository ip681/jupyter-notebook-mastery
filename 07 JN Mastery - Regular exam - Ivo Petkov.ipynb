## Visualizations

<b><font color='#8B0000'>TASK</font></b>

Create a new dataframe, called **"mt_ratio"**. It should contain **1 row** and **2 columns**. 

The dataframe should contain the ratio of the different types of content to all available content on Netlix. 

For example, if 81% of the content on Netflix is represented by movies, you should have a column with the type Movie and 0.81 content ratio.

**Tip**: There is a column in the dataframe which designates the type of the content.

```python
#YOUR CODE HERE
content_ratios = netflix_data['type'].value_counts(normalize=True)

mt_ratio = pd.DataFrame({'Content Type': content_ratios.index, 'Ratio': content_ratios.values})
mt_ratio['Ratio'] = mt_ratio['Ratio'].map('{:.2%}'.format)
mt_ratio
```

<b><font color='#8B0000'>TASK</font></b>

Create a **horizontal bar plot** which presents the relative share of the different types of content on Netflix (Movies and TV Series).

***REQUIREMENTS:***
- figure DPI should be **140**
 - figure size should be **(6.5, 2.5)**
 - use `#b20710` color for Movies, and `#221f1f` for TV Series
 - figure title should be **'Relative share of types of content on Netflix'**
 - create a figure **legend** in the **upper right** corner of the figure
 - no axis

```python
# #YOUR CODE HERE


content_ratios = netflix_data['type'].value_counts(normalize=True)
mt_ratio = pd.DataFrame({'Content Type': content_ratios.index, 'Ratio': content_ratios.values})

filtered_mt_ratio = mt_ratio[mt_ratio['Content Type'].isin(['Movie', 'TV Show'])]

colors = {'Movie': '#b20710', 'TV Show': '#221f1f'}

fig, ax = plt.subplots(figsize=(6.5, 2.5), dpi=140)

for idx, row in filtered_mt_ratio.iterrows():
    ax.barh(y=row['Content Type'], width=row['Ratio'], color=colors[row['Content Type']], label=row['Content Type'])

plt.title('Relative share of types of content on Netflix')
plt.legend(loc='upper right')
plt.axis('off')

plt.show()
```

<b><font color='#8B0000'>TASK</font></b>

The helper cell above create new columns which you can use to create a line plot.

Create a **line plot** which presents the content added on Netflix throughout the years.

***REQUIREMENTS:***


- figure size should be **(12, 5)**
- Movie line should be **colored** with `#b20710`,
- TV Show line must be **colored** with `#221f1f"`,
- y axis should be named **'Number of contributions'**
- figure title should be **'Contributions throughout the years'**
- area below every line should be **filled** with the respective color

```python
contributions_by_year = netflix_data.groupby(["year_added", "type"]).size().unstack(fill_value=0)

fig, ax = plt.subplots(figsize=(12, 5))

ax.plot(contributions_by_year.index, contributions_by_year["Movie"], label="Movies", color="#b20710")
ax.fill_between(contributions_by_year.index, contributions_by_year["Movie"], color="#b20710", alpha=0.3)
ax.plot(contributions_by_year.index, contributions_by_year["TV Show"], label="TV Shows", color="#221f1f")
ax.fill_between(contributions_by_year.index, contributions_by_year["TV Show"], color="#221f1f", alpha=0.3)
ax.set_ylabel("Number of contributions")
ax.set_title("Contributions throughout the years", fontsize=14)
ax.legend(loc="upper left")

plt.show()
```

```python
#YOUR CODE HERE
movies_data = netflix_data[netflix_data['type'] == 'Movie']
tv_shows_data = netflix_data[netflix_data['type'] == 'TV Show']

plt.figure(figsize=(12, 5))
plt.plot(movies_data['year_added'], range(len(movies_data)), label='Movies', color='#b20710', linewidth=2, alpha=0.7)
plt.plot(tv_shows_data['year_added'], range(len(tv_shows_data)), label='TV Shows', color='#221f1f', linewidth=2, alpha=0.7)
plt.fill_between(movies_data['year_added'], 0, len(movies_data), color='#b20710', alpha=0.2)
plt.fill_between(tv_shows_data['year_added'], 0, len(tv_shows_data), color='#221f1f', alpha=0.2)
plt.title('Contributions throughout the years')
plt.xlabel('Year')
plt.ylabel('Number of contributions')
plt.legend(loc='upper right')

plt.show()
```
