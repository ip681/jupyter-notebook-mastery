```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
plt.plot([1, 2, 3, 4])
plt.xlabel("something")
plt.ylabel("some numbers")
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])

plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], "ro")
# plt.axis((0, 6, 0, 20)) #xmin, xmax, ymin, ymax
plt.xlim(0, 6)
plt.ylim(0, 20)

plt.show()
```

```python
t = np.arange(0., 5., 0.2)

plt.plot(t, t, "r--", t, t**2, "bs", t, t**3, "g^")

plt.show()
```
